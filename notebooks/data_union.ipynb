```python
import os
import pandas as pd
```

# Data Union

### Absolute path where the data is stored

```python
data_path = os.path.abspath('../data')
```

### Data union process

Before starting with the deep analysis of the data, we are presented with the inconvenience of having the data we need to build our DataFrame in multiple '.csv' files.

We could have a DataFrame for each file, however this is not convenient for this particular case where we want to have a single DataFrame with all the graduates from 2006 to 2017.

With this in mind we proceed to create a function that as a result reads and concatenates all the '.csv' files we need to obtain our desired DataFrame.

```python
def data_union():
  data_files = os.listdir(data_path)
  arr_df = [pd.read_csv(f"{data_path}/{df}", sep=";") for df in data_files]
  df = pd.concat(arr_df, ignore_index=True)
  return df

data_union()
```

### Saving of the final DataFrame in 'CSV' format

Once the process of reading and concatenating the files is finished, we finally proceed to save this DataFrame in 'CSV' format to be able to use it in other notebooks.

```python
graduates_UBV_2006_2017 = data_union()

graduates_UBV_2006_2017.to_csv(f"{data_path}/graduates_UBV_2006_2017.csv")
```
